```python
%matplotlib inline
import pandas as pd
import io

url = 'https://raw.githubusercontent.com/Rdatasets/Rdatasets/master/csv/ggplot2/diamonds.csv'
df = pd.read_csv(url, index_col=0)

print("c) First 7 rows:")
print(df.head(7))

print("d) last 7 rows:")
tail = df.tail(7).to_string(header=False)
print(tail)
print()

print("e) descriptive statistics for numerical columns:")
num_stats = df.describe().to_string()
print(num_stats)
print()

print("f) descriptive statistics for categorical columns:")
category_stats = df.describe(include='object').to_string()
print(category_stats)
print()

print("g) unique category values:")
unique = df.select_dtypes(include='object').apply(pd.Series.unique)
print(unique)
print()

print("h) graph:")
df.hist(figsize=(20,10))




```
